# Label Encoding

```python
# Importing label encoder 
from sklearn import preprocessing 

# label_encoder object labels word labels. 
label_encoder = preprocessing.LabelEncoder() 
```

```python
# from above output we can see few object datatype features
# here we label all those objects- converting to int datatype

data['id']= label_encoder.fit_transform(data['id']) 
data['datetime']= label_encoder.fit_transform(data['datetime']) 
data['timezone']= label_encoder.fit_transform(data['timezone'])
data['destination']= label_encoder.fit_transform(data['destination']) 
data['product_id']= label_encoder.fit_transform(data['product_id'])
data['short_summary']= label_encoder.fit_transform(data['short_summary'])
data['long_summary']= label_encoder.fit_transform(data['long_summary'])
data['icon']= label_encoder.fit_transform(data['icon'])
data['cab_type']= label_encoder.fit_transform(data['cab_type'])
data['name']= label_encoder.fit_transform(data['name'])
data['source']= label_encoder.fit_transform(data['source'])
```

# Model prediction for 56 features

```python
import numpy as np
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

from sklearn.feature_selection import RFE
```

```python
#Creating model
reg = LinearRegression()
#Fitting training data
reg = reg.fit(X_train, y_train)
reg.score(X_train, y_train)
```

# sorting columns with accuracies

```python
y_size = rfecv.n_features_
# print(y_size)
dict_acc ={}
col = XX.columns
reg_single_col = []
reg_all_col = []

for i in range(y_size):
    X_col = XX.iloc[:,i:i+1]
    X_train, X_test, y_train, y_test = train_test_split(X_col, y, test_size = 0.3, random_state = 40)

    print("X_train shape : ",X_train.shape)
    print("y_train shape: ",y_train.shape)

    print("\nX_test shape : ",X_test.shape)
    print("y_test shape: ",y_test.shape)

    #Creating model
    reg1 = LinearRegression()
    #Fitting training data
    reg1 = reg1.fit(X_train, y_train)
    print("\nAccuracy : ",reg1.score(X_train, y_train), "  ---for column  ",i,end="\n\n")
    dict_acc[col[i]] = reg1.score(X_train, y_train)
    reg_single_col.append(reg1)
    
    
    #for all col without 1 column
    string = col[i]
    X_col = XX.drop([string],axis=1)
    X_train, X_test, y_train, y_test = train_test_split(X_col, y, test_size = 0.3, random_state = 40)

    print("X_train shape : ",X_train.shape)
    print("y_train shape: ",y_train.shape)

    print("\nX_test shape : ",X_test.shape)
    print("y_test shape: ",y_test.shape)

    #Creating model
    reg1 = LinearRegression()
    #Fitting training data
    reg1 = reg1.fit(X_train, y_train)
    print("\nAccuracy : ",reg1.score(X_train, y_train), "  ---for except column  ",i,end="\n\n")
    reg_all_col.append(reg1)
    
# dic = dict_acc.sorted()
print(dict_acc)
dic_ranking = sorted(dict_acc.items(), key=lambda x: x[1], reverse=True)
print(dic_ranking)
```
